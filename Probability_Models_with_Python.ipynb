```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
```

```python
print(rng.random())
print(rng.random(5))
print(rng.random([3,3]))
```

```python
# Generate a Binomial(50,0.2) random variable
def binomial(n,p):
  U=rng.random(n)
  return sum(U < p)

print(binomial(50,0.5))
```

```python
# random choice from array
def X ():
  A = [0.35 ,0.5 ,0.9 ,1]
  x = [1 ,2 ,3 ,4]
  j = 0
  U = rng . random ()
  while U > A [ j ]:
    j = j + 1
    return x [ j ]
```

```python
print(X())
# identisch mit
print(rng.choice([1,2,3,4],p=[0.35,0.15,0.4,0.1]))
```
